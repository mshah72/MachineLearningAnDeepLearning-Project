```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import time 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
```

```python
# Load the datasetata = "usedcars.csv"
train_data = 'test_data.csv'
train = pd.read_csv(train_data)

test_data = 'test_data.csv'
test = pd.read_csv(test_data)
```

```python
X_train = train.drop(['price'], axis=1)
y_train = train['price']
X_test = test.drop(['price'], axis=1)
y_test = test['price']
```

```python
# Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
start_rf = time.time()
rf.fit(X_train, y_train)
stop_rf = time.time()
y_pred_rf = rf.predict(X_test)

print(stop_rf - start_rf)

mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

# Print the results
print("Random Forest Regressor:")
print(f"MSE: {mse_rf:.2f}, RMSE: {rmse_rf:.2f}, R-squared: {r2_rf:.2f}, MAE: {mae_rf:.2f}")
```

```python
y1pred =  rf.predict(X_test) # predicting test data
y1_hat = rf.predict(X_train) # predicting train data
```

```python
# plotting trained price v/s predicted trained price
plt.scatter(y_train,y1_hat,s=0.5,c='b')
plt.plot(y_train,y_train,c='red')
plt.xlabel('trained y'); plt.ylabel('yhat')
plt.title("Plot of Train Data")
```

```python
# plotting tested price v/s predicted tested price
plt.scatter(y_test,y1pred,s=0.5,c='b')
plt.plot(y_test,y_test,c='r')
plt.xlabel('train y'); plt.ylabel('yhat')
plt.title("Plot of Test Data")
```
