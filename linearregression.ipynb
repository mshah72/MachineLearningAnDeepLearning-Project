```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import time 
import matplotlib.pyplot as plt
```

```python
# Load the datasetata = "usedcars.csv"
train_data = 'test_data.csv'
train = pd.read_csv(train_data)

test_data = 'test_data.csv'
test = pd.read_csv(test_data)
```

```python
X_train = train.drop(['price'], axis=1)
y_train = train['price']
X_test = test.drop(['price'], axis=1)
y_test = test['price']
```

```python
# Train the linear regression model
model = LinearRegression()
start_linear = time.time()
model.fit(X_train, y_train)
stop_linear = time.time()
y_pred_linear = model.predict(X_test)

print(stop_linear - start_linear)


mse_linear = mean_squared_error(y_test, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y_test, y_pred_linear)
mae_linear = mean_absolute_error(y_test, y_pred_linear)

# Print the results
print("\nLinear regression:")
print(f"MSE: {mse_linear:.2f}, RMSE: {rmse_linear:.2f}, R-squared: {r2_linear:.2f}, MAE: {mae_linear:.2f}")
```

```python
y1pred = model.predict(X_test) # predicting test data
y1_hat = model.predict(X_train) # predicting train data
```

```python
# plotting trained price v/s predicted trained price
plt.title("Plot of Train Data")
plt.scatter(y_train,y1_hat,s=0.5,c='b')
plt.plot(y_train,y_train,c='red')
plt.xlabel('trained y'); plt.ylabel('yhat')
```

```python
# plotting tested price v/s predicted tested price
plt.title("Plot of Test Data")
plt.scatter(y_test,y1pred,s=0.5,c='b')
plt.plot(y_test,y_test,c='r')
plt.xlabel('train y'); plt.ylabel('yhat')
```
