```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import time 
import matplotlib.pyplot as plt
```

```python
# Load the datasetata = "usedcars.csv"
train_data = 'test_data.csv'
train = pd.read_csv(train_data)

test_data = 'test_data.csv'
test = pd.read_csv(test_data)
```

```python
X_train = train.drop(['price'], axis=1)
y_train = train['price']
X_test = test.drop(['price'], axis=1)
y_test = test['price']
```

```python
# Decison Tree
dt = DecisionTreeRegressor(max_depth=10)
start_dt = time.time()
dt.fit(X_train, y_train)
stop_dt = time.time()
y_pred_dt = dt.predict(X_test)

print(stop_dt - start_dt)

mse_dt = mean_squared_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_test, y_pred_dt)
mae_dt = mean_absolute_error(y_test, y_pred_dt)

# Print the results
print("Decision Tree:")
print(f"MSE: {mse_dt:.2f}, RMSE: {rmse_dt:.2f}, R-squared: {r2_dt:.2f}, MAE: {mae_dt:.2f}")
```

```python
y1pred =  dt.predict(X_test) # predicting test data
y1_hat = dt.predict(X_train) # predicting train data
```

```python
# plotting trained price v/s predicted trained price
plt.scatter(y_train,y1_hat,s=0.5,c='b')
plt.plot(y_train,y_train,c='red')
plt.xlabel('trained y'); plt.ylabel('yhat')
plt.title("Plot of Train Data")
```

```python
# plotting tested price v/s predicted tested price
plt.scatter(y_test,y1pred,s=0.5,c='b')
plt.plot(y_test,y_test,c='r')
plt.xlabel('Actual Data'); plt.ylabel('Predicted Data')
plt.title("Decision Tree (with SelectKBest)")
```
