```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import time 
import matplotlib.pyplot as plt
```

```python
# Load the datasetata = "usedcars.csv"
train_data = 'test_data.csv'
train = pd.read_csv(train_data)

test_data = 'test_data.csv'
test = pd.read_csv(test_data)
```

```python
X_train = train.drop(['price'], axis=1)
y_train = train['price']
X_test = test.drop(['price'], axis=1)
y_test = test['price']
```

```python
# Ridge regression
ridge = Ridge(alpha=1)
start_ridge = time.time()
ridge.fit(X_train, y_train)
stop_ridge = time.time()
y_pred_ridge = ridge.predict(X_test)

print(stop_ridge - start_ridge)

mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)

# Print the results
print("Ridge regression:")
print(f"MSE: {mse_ridge:.2f}, RMSE: {rmse_ridge:.2f}, R-squared: {r2_ridge:.2f}, MAE: {mae_ridge:.2f}")
```

```python
y1pred =  ridge.predict(X_test) # predicting test data
y1_hat = ridge.predict(X_train) # predicting train data
```

```python
# plotting trained price v/s predicted trained price
plt.scatter(y_train,y1_hat,s=8,c='b')
plt.plot(y_train,y_train,c='red')
plt.xlabel('trained y'); plt.ylabel('yhat')
```

```python
# plotting tested price v/s predicted tested price
plt.scatter(y_test,y1pred,s=8,c='b')
plt.plot(y_test,y_test,c='r')
plt.xlabel('train y'); plt.ylabel('yhat')
```
